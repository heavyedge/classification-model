# Calibration scores

```python
from pathlib import Path
import pandas as pd

score_files = [
    Path("../../benchmarks/v1/CalibrationScores.sigmoid.csv"),
    Path("../../benchmarks/v1/CalibrationScores.isotonic.csv"),
    Path("../../benchmarks/v1/CalibrationScores.sigmoid_ovo.csv"),
    Path("../../benchmarks/v1/CalibrationScores.isotonic_ovo.csv"),
    Path("../../benchmarks/v1/CalibrationScores.temperature.csv"),
]

dfs = []
for path in score_files:
    _, method = path.stem.split(".")
    df = pd.read_csv(path).mean(axis=0)
    df["Method"] = method
    dfs.append(df)

pd.DataFrame(dfs).set_index("Method")
```
